## ¿Qué son las Redes Neuronales Artificiales?

Las redes neuronales artificiales están basadas en el funcionamiento de las redes de neuronas biológicas. Las neuronas que todos tenemos en nuestro cerebro están compuestas de dendritas, el soma y el axón: Las **dendritas** se encargan de **captar los impulsos nerviosos que emiten otras neuronas**. Estos impulsos, **se procesan en el soma** y **se transmiten a través del axón** que emite un impulso nervioso hacia las neuronas adyacentes.





### Neuronas artificiales

**Neurona:  unidad básica de procesamiento dentro de una red neuronal.**

Las neuronas artificiales se modelan de tal forma que imiten el comportamiento de una neurona cerebral. Tendrán unas ramificaciones y un núcleo o nodo. Habrá ramificaciones de entrada al nodo, que serán **las entradas de la neurona**, procedentes de otras neuronas seguramente.

Esta información se procesará en **un nodo**, y se generará una información de salida que se trasmitirá por las **ramificaciones de salida** a otras neuronas.

La imagen de una típica neurona artificial es la siguiente.




#### Una primera introducción al comportamiento de una neurona sería:





* $X_1$, $X_2$ son los valores de **entrada** desde el exterior de la neurona.
* Dentro de la neurona se realiza un **cálculo interno**.
* $Y_1$ es el valor de **salida**.
* ___b___ es el término independiente, también se llama **sesgo** o **bias**.
* $W_1$, $W_2$ son los **parámetros (pesos) de nuestra neurona**.



Todavía el proceso es similar al de la ___regresión lineal múltiple___. Para que se pueda diferenciar añadiremos un concepto más:  **función de activación**.  

Lo que podemos hacer es añadir **deformaciones no lineales** que nos permiten tener **una combinación incluso no-lineal de las entradas** a nuestra neurona. Es decir, **transformamos la suma ponderada de los valores entrantes** con estas funciones.

### Red Neuronal simple y Red Neuronal profunda

Como se puede observar en esta representaión de las redes neuronales artificiales, las neuronas de color **rojo** no tienen ramificaciones de entrada. Son información que vamos a dar a las neuronas **desde el exterior**, o mejor dicho los **valores que se vienen dado al inico** de una observación considerando como un **estímulo inicial**. Estas neuronas forman **la capa de entrada (Input Layer)**.

Las neuronas de color **azul** tienen ramificaciones de salida que no están conectadas a otras neuronas. Son **la información de salida** de la red neuronal al espacio exterior o el **estímulo final**. Estas neuronas forman **la capa de salida (Output Layer)**.

El resto de las neuronas de color **amarillo** que están interconectadas forman las capas interiores de la red. Dependiendo del **número de capas ocultas (Hidden Layers)** podemos hablar de una red neuronal **simple** o una red neuronal **profunda**.

### Valores de salida y pesos de las redes neuronales

Las respuestas calculadas por las neuronas o los valores de salida pueden ser:
* **continuos**, como el precio de una compra determinada o una vivienda!
* **categóricos**, como el producto que más se venderá el próximo año o la especie de una flor!.
* **binarios**, como por ejemplo si una persona padece de una enfermedad o no. (_el caso binario es un caso particular donde tenemos dos categorías_).

Para poder llevar a cabo el procesado, cada conexión entre las neuronas llamada ***sinapsis*** tendrá asignada un valor o un **peso** que equivale a la **fuerza** de la señal que se trasmite a otras. El ajuste de estos pesos para tener una red neuronal que haga lo que queremos es fundamental.

Existen **distintos métodos para ajustar estos pesos**, como **el descenso de gradiente (Gradient Descent)** y **el Backpropagation**.

## Funciones de activación

Como se ha observado anteriormente, primero se calcula **una suma ponderada** usando los pesos y las variables de entrada, y después **se le aplica una función** sobre dicha combinación. Las **funciones de activación** realmente son la manera de **considerar y trasmitir esta información** por las conexiones de salida.

Las funciones de activación se utilizan para aplicar, añadir o dar una **no linealidad** al modelo y hacen que la red sea capaz de **resolver problemas más complejos**.  

**`La función de activación decide si una neurona debe activarse o no`**. Esto significa que **decidirá si la participación de la neurona en la red es importante o no de cara al proceso de predicción**. Vamos a repasar las cuatro familias de funciones de activación más usadas en redes neuronales artificiales:

### **Función escalón (threshold - step function)**

Como vemos en la figura de abajo, en el eje horizontal está representado el valor ya ponderado ($x$) de las entradas ($x_i$) y sus respectivos pesos ($w_i$), y en el eje vertical tenemos el valor de la función escalón ($y$). Esta función propaga un **0** si el valor de **_x_ es negativo**, o un **1** si es **positivo**.  

**No hay casos intermedios**, y sirve para **clasificar de forma muy estricta**. Es la función **más rígida**.



### **Función sigmoide**

Al igual que la función escalón, existe una división entre los valores negativos y positivos de la suma ponderada ($x$), pero la función sigmoide **no es tan estricta** porque el cambio no se hace de golpe y se realiza de manera suave. **Se usa además en la regresión logística**, una de las técnicas más usadas en Machine Learning.

Como vemos en la figura de abajo, la función no tiene esquinas y es más suave (**en términos matemáticos, es derivable**). Cuanto más positivo es el valor de ($x$) más nos acercamos a 1 y por el contrario, cuanto más negativo es ($x$) más nos acercamos a 0.

Esta función es **muy útil en la capa salida al final de la red neuronal**, no solo para clasificar con valores categóricos, sino también para intentar **predecir las probabilidades de pertenencia a cada categoría**, donde sabemos que la probabilidad de un suceso imposible es 0 y la de un suceso seguro es 1.



### **Función rectificadora (ReLU)**

**Rectified Linear Unit** o la función ReLU **no activa como _Sigmoide_ a casi todas las neuronas de la red**. Vemos en la figura de abajo que su valor es **0** para cualquier valor negativo de ($x$).  Si ($x$) es positivo la función **retorna el propio valor de ($x$)**.

Por lo tanto, se obvian todas aquellas entradas que una vez ponderadas tienen un valor negativo y nos quedamos con el valor exacto de las positivas. Esto hace que nuestra red sea **computacionalmente mucho más eficiente** comparando con casos anteriores.



### **Función tangente hiperbólica (tanh)**

Esta función es **muy parecida a la función _sigmoide_**, pero en este caso, como vemos en la figura de abajo, no empieza en 0 y termina en 1, sino que **empieza en -1 y termina en 1**. La salida de esta función está **centrada en cero** y por lo tanto, **activaría los valores de salida cuando son rotundamente negativos o rotundamente positivos**.



### Configuración de la red neuronal

Ya tenemos los **ingrediente básicos** para configurar una red neuronal.
* Una **capa de neuronas de entrada**, donde pasaremos información a la red a modo de **estímulos**.
* una **capa de neuronas oculta**, que se encargará del **procesado** de la información proporcionada por la capa de entrada.
* una **capa de salida** que procesará la información proporcionada por la capa oculta para **obtener el resultado final** de la red neuronal.




En esta imagen vemos que en la capa oculta usamos una función rectificadora (_ReLU_) y en la de salida usamos una función _sigmoide_. **Podemos usar cualquier otro tipo de función que mejor se ajuste al problema** que pretendemos resolver con nuestra red neuronal.  

La elección de una función de activación un tipo u otro dependerá de varios factores como:  

#### 1. Variable de salida:  
Uno de los aspectos claves a la hora de elegir estas funciones es **el rango de valores que queremos en la salida** de la neurona:
- Si **no hay restricciones** en el rango, utilizaremos la función de identidad (**identity**) que propagará en la salida el valor de entrada.  
- Si la restricción es únicamente que el valor de salida **debe ser positivo**, sin límite de cantidad, elegiremos una función rectificadora (**ReLU**).    
- Si queremos una **salida binaria** como 1 y 0 para clasificar, elegimos por ejemplo la función escalón (**threshold**).  

#### 2. Algoritmo de aprendizaje:  
Otro factor decisivo en el momento de seleccionar el tipo de la función de activación es la tipología del algoritmo que se aplica a los datos para ajustar los pesoso y los parámetros de la red neuronal, con el fin de llegar a un conjunto de valores óptimos que minimicen los errores:
- Puede que usemos un algoritmo de aprendizaje que **hace uso de derivadas** (por ejemplo como **Gradient Descent**). Entonces utilizaremos **funciones de activación donde la derivada esté definida en todo el intervalo**. Por ejemplo se puede optar por la **función de sigmoide** o la **tangente hiperbólica**, teniendo en cuenta que son **más costosas en términos computacionales**.

Definimos los datos del modelo con todas las variables de entrada **menos la variable éticamente incorrecta**


```python
X_multiple = df_boston.drop(['target', 'B'], axis='columns')
y_multiple = df_boston['target']
```

A pesar de que es una práctica común y recomendada, esta vez no modificamos la *[estructura de datos](https://scikit-learn.org/stable/glossary.html#term-array-like)* antes de pasarlo al modelo.

```python
# Dividir el dataset en Training y Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_multiple, y_multiple, test_size=0.2, random_state=222)
```

#### Aplicar el modelo de regresión no-lineal con “redes neuronales artificiales”: **Neural Network Regression**

```python
from sklearn.neural_network import MLPRegressor

red_neuronal = MLPRegressor(random_state=222)
```

```python
# Ajustamos el algoritmo al conjunto de datos de entrenamiento

red_neuronal.fit(X_train, y_train)
```

```python
# lanzamos las predicciones (para el conjunto de test que son los datos no vistos por el algoritmo en la fase de modelización)

y_pred = red_neuronal.predict(X_test)
```

#### Evaluación del modelo a través de las métricas de regresión

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# Métricas para evaluar la calidad del modelo
print('Mean Absolute Error:', mean_absolute_error(y_test, y_pred))
print('Mean Absolute Percentage Error:', mean_absolute_percentage_error(y_test, y_pred)*100)
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, y_pred)))
print('R^2 coefficient of determination:', r2_score(y_test, y_pred))
```

A primera vista vemos que obtenemos una **pésima precisión con el modelo por defecto**, comparando con métodos como *Regresión Forest* o *Decision Trees* que se desarrollaron en sesiones anteriores.

* ``hidden_layer_sizes``: tupla, longitud = n_capas - 2, predeterminado = (100,). El i-ésimo elemento representa el número de neuronas en la i-ésima capa oculta.

* ``alpha``: float, predeterminado = 0.0001. Parámetro de penalización L2 (término de regularización).
* ``tol``: float, predeterminado = 1e-4. Tolerancia para la optimización. Cuando la pérdida o la puntuación no mejoran al menos ``tol`` en ``n_iter_no_change`` iteraciones consecutivas se considera que se ha alcanzado la convergencia y se detiene el entrenamiento.

### Grid Search

Una de los métodos más usados en la búsqueda de ajustes óptimos es el método **GridSearchCV** que permite buscar los hiperparámetros dentro de unos rengos determinados para obtener el mejor modelo que se puede ajustar.

En este caso necesitamos en concreto **generar varias combinaciones de tamaños de capas ocultas** que considerar en el _Grid_.  

Para ello podemos usar el módulo **itertools**. Este módulo consiste en una colección de herramientas para manejar ___iteradores (iterators)___. Los iteradores son estructuras de datos que se consideran como **objetos iterables** y por lo tanto se pueden **usar en un bucle _for_**. El iterador más común en Python es la lista.

```python
# Ejemplo de un iterador que va hasta infinito
import itertools

iterador = itertools.count(start=0, step=2)

for k in iterador:
    if k < 10:
        print(k)
    else:
        break
```

```python
# Ejemplo de un iterador que se repite
import itertools

iterador = itertools.repeat('Neural Network Regression', times=3)

for k in iterador:
    print (k)
```

Ahora vemos cómo podemos conseguir un grupo de **diferentes conjuntos de tamaños de las capas ocultas** (por ejemplo _(100, 120, 100)_)  que se pueda iterar dentro del método _GridSearchCV_

```python
# Primero vemos un ejemplo de pares de valores que pueden tomar'A' o 'B' como sus valores individuales
import itertools

for x in itertools.combinations(['A','B'], 2):
    print(x)
```

```python
# En caso de que queremos considerar el orden
import itertools

for x in itertools.permutations(['A','B'], 2):
    print(x)
```

```python
# Si queremos combinaciones que admiten repeticiones tenemos:
import itertools

for x in itertools.combinations_with_replacement(['A','B'], 2):
    print(x)
```

```python
# Todos los pares posibles del conjunto de valores 'A' y 'B' sería:
import itertools

for x in set(list(itertools.permutations(['A','B'],2)) + list(itertools.combinations_with_replacement(['A','B'],2))):
    print(x)
```

```python
# La forma más óptima sería usar el producto cartesiano del conjunto de los valores permitidos consigo msmo
import itertools

for x in itertools.product(['A','B'], repeat=2):
    print(x)
```

A parte de comparar y revisar los resultados finales del modelo, deberíamos poder analizar la complejidad y la capacidad de aprendizaje del modelo para estudiar los posibles efectos de varianza y sobre todo posibles __Overfittings__ que puede llegar a tener nuestra red neuronal.

Ahora vamos a sacar una grafica tipo ___Complexity Curve___ al igual que algunos otros modelos desarrollados en sesiones anteriores como _Random Forest_ o _Decision Trees_.

Probamos si podemos analizar la complejidad del modelo, **aumentando un parámetro como el número de las neuronas en la capa oculta** en nuestra red neuronal:

```python
# Consideramos un rango para asignar el hiperparámetro
hiper_param = np.arange(10,1201,10)

# Generamos previamente los vectores necesarios para ir calculando y guardando el rendimiento
train_r2 = np.zeros(hiper_param.size)
test_r2 = np.zeros(hiper_param.size)

for i in range(hiper_param.size):
    # Generamos un modelo para cada hiperparámetro, lo entrenamos y calculamos el R_cuadrado sobre datos de train y de test
    mod_red = MLPRegressor(hidden_layer_sizes=hiper_param[i],
                              random_state=222)
    mod_red.fit(X_train, y_train)
    train_r2[i] = r2_score(y_train, mod_red.predict(X_train))
    test_r2[i] = r2_score(y_test, mod_red.predict(X_test))

print("El mejor valor de n_estimator podría ser =", hiper_param[np.argmax(test_r2)],
      " que consigue un R2 =", max(test_r2))

# Graficamos el R_cuadrado de training versus de test
fig = plt.figure(figsize=(20,7))
plt.plot(hiper_param, test_r2, linewidth=3, label='Test R^2')
plt.plot(hiper_param, train_r2, linewidth=3, label='Train R^2')
plt.plot(hiper_param[np.argmax(test_r2)], max(test_r2),
        marker='o', color = "red", label="max R^2")
plt.xticks(hiper_param)
plt.xlabel('Complejidad (tamaño de la capa oculta)')
plt.ylabel('R2')
plt.legend(loc = 'lower right')
plt.show()

```

Se puede observar que los modelos conseguidos incluso con un número muy elevado de neuronas en la capa oculta, sufren de ___Underfitting___ más que tener un sobre ajuste a los datos de entrenamineto.  

Esto quiere decir que en caso de las redes neuronales **no podemos sacar tan fácilmente una gráfica muy concreta de complejidad por no disponer de un único parámetro** que se pueda estudiar de la misma manera que otros algoritmos.

A cambio, lo que sí podemos hacer es realizar un estudio sobre el ritmo y la evolución de su capacidad de aprendizaje. Por ello definimos una nueva función al igual que otros modelos para sacar la gráfica de ___Learning Curve___.

```python
# Definimos una función que toma el valor de los hiperparámetros y
# dibuja la evolución del rendimiento del modelo para el conjunto de training y de test:
def curva_aprendizaje_red(capas_ocultas: tuple=(100,),
                         funcion_activacion: str='relu',
                         optimizador: str='adam'):
    """Función para sacar la gráfica de Learning Curve a partir de los hiperparámetros de la red neuronal"""

    # Generar la estructura del modelo
    modelo_red = MLPRegressor(hidden_layer_sizes=capas_ocultas,
                              activation=funcion_activacion,
                              solver=optimizador,
                              random_state=222)
    # Crear un listado de los tamaños de subconjuntos de datos de entrenamiento
    num_samples = np.linspace(10,X_train.shape[0]).astype(int)

    # Generamos previamente los vectores necesarios para ir calculando y guardando el rendimiento de los modelos
    train_R2 = np.zeros(num_samples.size)
    test_R2 = np.zeros(num_samples.size)

    for i in range(num_samples.size):
        # Generamos un modelo para cada subconjunto de datos de entrenamiento y lo ajustamos
        modelo_red.fit(X_train[:num_samples[i]], y_train[:num_samples[i]])

        # Calculamos el R_cuadrado sobre datos de train y de test
        train_R2[i] = r2_score(y_train[:num_samples[i]], modelo_red.predict(X_train[:num_samples[i]]))
        test_R2[i] = r2_score(y_test, modelo_red.predict(X_test))

    print("La diferencia de R^2 entre training y test es= ", train_R2[-1]-test_R2[-1])

    # Graficamos el R_cuadrado de training versus de test
    fig = plt.figure(figsize=(20,7))
    plt.plot(num_samples, test_R2, label = 'Test R^2')
    plt.plot(num_samples, train_R2, label = 'Train R^2')
    plt.title('Curva de aprendizaje (Learning Curve) de la red neuronal')
    plt.xlabel('Tamaño de entrenamiento (Training size)')
    plt.ylabel('R2')
    plt.legend(loc = 'lower right')
    plt.show()
```

```python
# Aplicamos la función con los valores por defecto que corresponden al modelo inicial
curva_aprendizaje_red()
```

Se puede apreciar que el modelo sigue sufriendo de un ___Underfitting___ más que estar sobre ajustado a los datos de training.

Ahora sacamos el _Learning Curve_ para el segundo modelo.

```python
curva_aprendizaje_red(capas_ocultas=(120, 120, 120),
                      optimizador='lbfgs')
```

En esta gráfica vemos que la evolución del $R^2$ no indica un riesgo de _Overfitting_ y la calidad de los modelos ajustados se mejora bastanta.  

Procedemos a graficar el _Learning Curve_ de la red neuronal con los hiperparámetros calculados por _Grid Search_.

```python
curva_aprendizaje_red(capas_ocultas=(100, 120, 80),
                      funcion_activacion='tanh',
                      optimizador='lbfgs')
```

En esta caso vemos que **las dos líneas de evolución de $R^2$ se convergen** al final, además llegan a tener valores significativos que indican la buena calidad del modelo ajustado con los hiperparámetros óptimos.
